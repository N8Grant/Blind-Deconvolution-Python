```python
from pylab.image.restoration.blind_deconv import blind_deconvolution
import PIL
import matplotlib.pyplot as plt
import numpy as np
from scipy.signal import convolve2d
```

## Original Image

```python
image_original = PIL.Image.open('./examples/old_pic.jpeg')
image_original_np = np.asarray(PIL.ImageOps.grayscale(image_original))
plt.imshow(image_original_np)
```

## Add a gaussian blur

```python
gaussian = np.array([[1,2,1],
            [2,4,2],
            [1,2,1]])*(1/16)
blurred_image = convolve2d(image_original_np, gaussian)
```

```python
plt.imshow(blurred_image)
```

## Use Blind Deconvolution to Remove Blur

```python
restored_image = blind_deconvolution(blurred_image)
```

```python
plt.imshow(restored_image)
```
